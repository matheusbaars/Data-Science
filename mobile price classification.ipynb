```python
features = ['battery_power', 'blue', 'fc', 'int_memory', 'mobile_wt', 'pc', 'px_height', 'px_width', 'ram',
           'sc_h', 'talk_time', 'three_g', 'touch_screen']
```

```python
Data = dataset[features]
```

```python
Target = dataset['price_range']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(Data, Target, test_size=0.2, random_state=100)
```

```python
dtc = tree.DecisionTreeClassifier()
dtc = dtc.fit(X_train, Y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test,y_pred)*100
print(accuracy)
```

## encontrando o melhor nÃ­vel

```python
for i in range(1, 50):
    dtc = tree.DecisionTreeClassifier(max_depth=i, min_samples_leaf=5, random_state=100)
    dtc = dtc.fit(X_train, Y_train)
    y_pred = dtc.predict(X_test)
    accuracy = accuracy_score(Y_test,y_pred)*100
    print('NÃ­vel = {} | AcurÃ¡cia = {}'.format(i, accuracy))
```
